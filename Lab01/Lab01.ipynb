# Importando bibliotecas

```python
import io
import requests
import json
import pandas as pd
import xml.etree.ElementTree as et
```

# Tarefa 1
Fonte escolhida: Prevalence of current tobacco use among adolescents (13-15 years) (%)

Link: https://apps.who.int/gho/athena/api/GHO/MDG_0000000015

```python
url = "https://apps.who.int/gho/athena/api/GHO/WHOSIS_000013"
```

## Json

```python
js = requests.get(url + "?format=json").content
print(js.decode("utf-8"))
```

## CSV

```python
data = requests.get(url + "?format=csv").content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
csv.head(10)
```

## XML

```python
data = requests.get(url).content
x = et.parse(io.StringIO(data.decode("utf-8")))

print(et.tostring(x.getroot()))
```
